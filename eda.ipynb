# Exploratory Data Analysis

## This notebook focuses on exploratory data analysis to identify trends, patterns, and relationships that inform hypothesis formulation and model development.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('cleaned_superstore.csv')
```

## Univariate Analysis

```python
plt.figure(figsize=(6, 4))
sns.histplot(df['Sales'], bins=50, kde=True)
plt.title("Distribution of Sales")
plt.show()
```

Sales values are right-skewed, indicating a large number of low-value transactions and a small number of high-value orders.

```python
plt.figure(figsize=(6, 4))
sns.histplot(df['Profit'],bins=50,kde=True)
plt.title("Distribution of Profit")
plt.show()
```

The profit distribution contains both positive and negative values, suggesting the presence of loss-making transactions.

```python
plt.figure(figsize=(6,4))
sns.histplot(df['Discount'],bins=20,kde=True)
plt.title('Distribution of Discount')
plt.show()
```

Discounts are concentrated at specific discrete levels, indicating predefined promotional strategies rather than continuous discounting.

## Bivariate analysis

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='Discount', y='Profit',data=df,alpha=0.5)
plt.title('Discount vs Profit')
plt.show()
```

A negative relationship is observed between discount and profit, where higher discounts often correspond to reduced or negative profits.

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='Sales',y='Profit',data=df,alpha=0.5)
plt.title("Sales vs Profit")
plt.show()
```

Higher sales are generally associated with higher profit, several high-sales transactions result in losses, suggesting the influence of discounts or cost structures.

## Categorical Comparisons

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='Category',y='Profit',data=df)
plt.title('Profit Distribution by Category')
plt.show()
```

Profitability varies significantly across categories, indicating that some product categories are more susceptible to losses.

```python
plt.figure(figsize=(6,4))
sns.boxplot(x="Region",y="Profit",data=df)
plt.title("Profit Distribution by Region")
plt.show()
```

Regional differences in profit suggest the potential influence of logistics, customer behavior, or regional pricing strategies.

## Key EDA Findings:
* Sales and profit distributions are highly skewed.
* Discounts appear to have a negative association with profit.
* High sales do not always guarantee profitability.
* Profitability differs across categories and regions.
