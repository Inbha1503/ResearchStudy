# Predictive modeling

This section develops a regression-based predictive model to quantify the influence of sales, discount, and quantity on profit.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
df=pd.read_csv('cleaned_superstore.csv')
```

```python
features = ['Sales', 'Discount', 'Quantity']
target = 'Profit'

x = df[features]
y = df[target]
```

Predictor variables were selected based on domain relevance and insights obtained from exploratory analysis.

```python
x.fillna(0)
y.fillna(0)
```

Missing values were handled

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

The dataset was split into training and testing sets to evaluate model generalization.

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(mae, r2)
```

The model explains a substantial portion of variance in profit, as indicated by the R² score. The MAE suggests reasonable prediction accuracy given the variability of retail profits.

```python
coefficients = pd.Series(model.coef_, index=features)
print(coefficients)
```

* Discount shows a strong negative coefficient, confirming that higher discounts significantly reduce profit.
* Sales has a positive coefficient, indicating that increased revenue contributes positively to profit.
* Quantity shows a comparatively smaller impact.

## **Predictive Modeling Conclusion:**
Regression analysis confirms that **discount level is the most influential factor affecting profit, with a strong negative effect**. These findings align with both exploratory analysis and hypothesis testing, reinforcing the conclusion that **aggressive discounting negatively impacts retail profitability.**
