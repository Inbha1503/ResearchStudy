```python
df['Order Date']=pd.to_datetime(df['Order Date'])
df['Ship Date']=pd.to_datetime(df['Ship Date'])
```

Date columns were converted to datetime format

```python
df['Order Year']=df['Order Date'].dt.year
df['Order Month']=df['Order Date'].dt.month
df['Profit Margin']=df['Profit']/df['Sales']
```

New columns were added to better analysation of patterns.

```python
df[['Sales','Profit','Discount','Quantity']].describe()
```

Descriptive statistics reveal significant variability in sales and profit, indicating the presence of both high-value and loss-making transactions.
